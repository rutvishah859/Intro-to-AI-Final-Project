```python
df = pd.read_csv('Age_Vs_Crime.csv')
df = df.set_index([pd.Index(['Pop. of Ages 0-14', 'Crime Rate of B&E', 'Num. of Schools'])])
df = df.T
df['Pop. of Ages 0-14'] = df['Pop. of Ages 0-14'].astype(str).str.replace(',', '').astype(float)
df['Crime Rate of B&E'] = df['Crime Rate of B&E'].astype(str).str.replace(',', '').astype(float)
df['Num. of Schools'] = df['Num. of Schools'].astype(str).str.replace(',', '').astype(float)
df
```

```python
ssd = []
K = range(1, 50)
for k in K:
   cluster = KMeans(n_clusters=k).fit(df)
   ssd.append(cluster.inertia_)
```

```python
plt.plot(K, ssd, 'bx-')
plt.xlabel('k')
plt.ylabel('sum of squared distances')
plt.title('Elbow Method for Optimal K')
```

```python
sil = []
K_sil = range(2, 50)
for k in K_sil:
  cluster = KMeans(n_clusters=k).fit(df)
  labels = cluster.labels_
  sil.append(silhouette_score(df, labels, metric='euclidean'))
```

```python
plt.plot(K_sil, sil, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method for Optimal K')
```

Pick k = 5 (optimal k according to elbow method). There is also a peak at k = 5, as observed with the silhouette method.

```python
#number of clusters
k = 5
#run K-Means clustering algorithm
cluster = KMeans(n_clusters=k).fit(df)
#prints the counts of neighbourhoods assigned to diff. clusters
print(collections.Counter(cluster.labels_))
```

```python
#labels 
labels = cluster.labels_
results = pd.DataFrame([df.index, labels]).T
results
```

```python
#joins the table with all the headings together
df['Score'] = df.apply(lambda row: row['Pop. of Ages 0-14']/row['Crime Rate of B&E']*row['Num. of Schools'], axis=1)
df['Cluster'] = results[1].values
df
```

```python
clusters = df.groupby('Cluster')
print(clusters['Score'].mean())
clusters = clusters['Score'].mean()
```

The best cluster for our goal is Cluster 4 with the highest mean score, followed by cluster 3, 2, 1 and 0 (assending order)
